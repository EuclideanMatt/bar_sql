```python
dfs = [dram_15, dram_16, dram_17, dram_18, dram_19, dram_20, dram_21]

def clean_dataframe(df):
    # Lowercase values in all columns
    df = df.apply(lambda col: col.map(lambda x: x.lower() if isinstance(x, str) else x))

    # Lowercase & remove spaces in headers
    df.columns = df.columns.str.lower().str.replace(' ', '_')

    # Removing '$' and '-' from monetary features
    columns_to_clean = ['gross_sales', 'discounts']
    if all(col in df.columns for col in columns_to_clean):
        df[columns_to_clean] = df[columns_to_clean].replace({'\$': '', '-': ''}, regex=True)
        df[columns_to_clean] = df[columns_to_clean].apply(pd.to_numeric, errors='coerce')

    # Drop specified columns
    columns_to_drop = ['unit', 'count', 'net_sales', 'modifiers_applied', 'tax', 'dining_option', 'time_zone']
    df = df.drop(columns=[col for col in columns_to_drop if col in df.columns], errors='ignore')

    return df

# Apply the cleaning function to each dataframe in the list
for i in range(len(dfs)):
    dfs[i] = clean_dataframe(dfs[i])


```

```python
dram_15 = clean_dataframe(dram_15)
dram_16 = clean_dataframe(dram_16)
dram_17 = clean_dataframe(dram_17)
dram_18 = clean_dataframe(dram_18)
dram_19 = clean_dataframe(dram_19)
dram_20 = clean_dataframe(dram_20)
dram_21 = clean_dataframe(dram_21)
```

```python
# creating our database 

conn = sqlite3.connect(r"C:\Users\Epicm\Desktop\data_bases\dbs\planes\dram_db.db")
cursor = conn.cursor()
```

```python
# Define the function to create a table and insert data
def create_table_and_insert_data(df, table_name):
    # Create table with a dynamic name
    cursor.execute(f'''
    CREATE TABLE IF NOT EXISTS {table_name} (
        date TEXT,
        time TEXT,
        category TEXT,
        item TEXT,
        qty INTEGER,
        price_point_name TEXT,
        sku TEXT,
        gross_sales FLOAT,
        discounts FLOAT,
        transaction_id TEXT,
        payment_id TEXT,
        device_name TEXT,
        notes TEXT,
        details TEXT,
        event_type TEXT,
        location TEXT,
        customer_id TEXT,
        customer_name TEXT,
        customer_reference_id TEXT
    )
    ''')

    # Insert data into the table
    for _, row in df.iterrows():
        cursor.execute(f'''
        INSERT INTO {table_name} (date, time, category, item, qty, price_point_name, sku, gross_sales, discounts, transaction_id, payment_id, device_name, notes, details, event_type, location, customer_id, customer_name, customer_reference_id)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        ''', tuple(row))
    
    conn.commit()

# List of cleaned dataframes and their corresponding table names
cleaned_dfs = {
    'dram_15': dram_15,
    'dram_16': dram_16,
    'dram_17': dram_17,
    'dram_18': dram_18,
    'dram_19': dram_19,
    'dram_20': dram_20,
    'dram_21': dram_21
}

# Iterate through each dataframe and create table and insert data
for table_name, df in cleaned_dfs.items():
    create_table_and_insert_data(df, table_name)

```

```python
# now I will create my own net_sales column to ensure accuracy 

conn = sqlite3.connect(r"C:\Users\Epicm\Desktop\data_bases\dbs\planes\dram_db.db")
cursor = conn.cursor()
```

```python
def add_net_sales_column():
    # Retrieve the names of all tables in the database
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = cursor.fetchall()

    for table in tables:
        table_name = table[0]
        
        # Add the 'net_sales' column if it doesn't already exist
        cursor.execute(f"ALTER TABLE {table_name} ADD COLUMN net_sales FLOAT;")

        # Update the 'net_sales' column with the calculated value
        cursor.execute(f'''
            UPDATE {table_name}
            SET net_sales = gross_sales - discounts
            WHERE gross_sales IS NOT NULL AND discounts IS NOT NULL;
        ''')

    # Commit changes to the database
    conn.commit()

# Call the function to add the net_sales column to all tables
add_net_sales_column()

# Close the database connection
conn.close()
```
